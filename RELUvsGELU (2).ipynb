## GELU vs ReLU – How AI Filters Thoughts

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

# Step 1: Generate input values from -3 to 3
x = torch.linspace(-3, 3, 100)

# Step 2: Use built-in activation functions
y_gelu = F.gelu(x)         # GELU is smooth and gentle
y_relu = F.relu(x)         # ReLU cuts off negatives

# Step 3: Plot both
plt.figure(figsize=(8, 3))

plt.subplot(1, 2, 1)
plt.plot(x.numpy(), y_gelu.numpy())
plt.title("GELU – Gentle Filter")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x.numpy(), y_relu.numpy())
plt.title("ReLU – Harsh Filter")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)

plt.tight_layout()
plt.show()
```

** Left: GELU – Gentle Filter**
Allows small/negative inputs to pass softly

Helps GPT learn from subtle, uncertain signals

Feels more "human" and thoughtful

** Right: ReLU – Harsh Filter**
Completely blocks negative inputs

Simpler and faster, but loses weak signals

Not ideal for complex language models
